## MatplotLib 

Matplotlib is a otting libirary for the python programming language and its numerica; mathematics extension Numpy, it provides an object-oriented API for embedding plots into application using general purpose GUI toolkit like Tkinter,Wxpython,Qt or GTK+

some of the major pros of matplotlib are :


generally easy to get started for simple plots 
support  for customer labes and texts 
great control of every element in a figure 
high quality output in many formats 
very customized in general

```python
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as  np
```

```python
x=np.arange(0,10)
y=np.arange(1,11)
```

```python
a=np.arange(20,30)
b=np.arange(40,50)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d diagram')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.plot(2,2,3)
plt.plot(a,b,'r')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title('matplotlib demo')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,8*np.pi,0.5)
a=np.sin(x)
b=np.cos(x)


plt.subplot(2,1,1)
plt.plot(x,b,'g')
plt.title('cos')




plt.subplot(2,1,2)
plt.plot(x,a)
plt.title("sin")
```

```python
## bar plot
x=[2,4,6,8,10]
y=[1,3,5,7,9]
a=[41,52,23,67,12]
b=[20,50,32,21,34]
plt.bar(x,y,align='center')
plt.bar(a,b,color='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
#histogram
a=np.array([12,22,34,54,52,51,45,32,14,43,45,24,28])
plt.hist(a,bins=20)
plt.title('Histogram')
plt.show()
```

```python
## box plot
a=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(a,vert=True,patch_artist=True)
```

```python
#pie chart
labels='ankita','madhukar','sangam','mangala'
sizes=[24,60,27,55]
colors=['blue','red','green','lightskyblue']
explode=(0.2,0,0,0)

plt.pie(sizes,explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=False)
plt.show()
```
