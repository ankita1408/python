#### correlation with heatmap
a correlation heatmap uses colored cells,typically in monochromatic scale to show a 2d correlation matrix(table) betweentwo discrete dimensions or event types it is very important in feature selection

```python
df.corr()
```

```python
df.dtypes
```

```python
df.info
```

```python
sns.heatmap(df.corr())
```

### joinplot
a join plot allows to study the relationship between 2 numeric variable.
the centeral chart display their correlation.it is usuallya scatterplot a hebin plot a2d histogram or 2d density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## pair plot
a pair plot is also call scatterplot .In which one in the same data row is matched with another variables value like this:pair plot are just elaboration on this showing all variables paired withh all the vairbles

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

# dist plot
dist plot helps us to check the distribuation of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
